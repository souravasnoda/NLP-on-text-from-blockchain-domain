LOADING DATA

```python
df = pd.read_csv('/content/root2ai - Data.csv')
```

DATA PREPROCESSING

```python
df.isnull().sum()
data = df.dropna(axis=0)
```

```python
x = data.iloc[:,0].values
target=  data.iloc[:,-1].values
print(x)
```

```python
from sklearn.preprocessing import LabelEncoder,OneHotEncoder
```

```python

encoder = LabelEncoder()
y=  encoder.fit_transform(target)


print(y)
```

CONVERTING ALL WORDS INTO LOWERCASE

```python
processed = data['Text'].str.lower()
print(processed)
```

REMOVING STOPWORDS

```python
from nltk.corpus import stopwords
stop_words = set(stopwords.words('english'))
processed = processed.apply(lambda x: ' '.join(term for term in x.split() if term not in stop_words))
```

APPLYING STEMMING

```python
ps = nltk.PorterStemmer()
processed = processed.apply(lambda x:' '.join(ps.stem(term) for term in x.split()))
```

TOKENIZATION

```python
nltk.download('punkt')


from nltk.tokenize import word_tokenize

all_words = []
for text in processed:
  word = word_tokenize(text)
  for w in word:
    all_words.append(w)

all_words = nltk.FreqDist(all_words)


print('number of words: {}'.format(len(all_words)))
print('most common words: {}'.format(all_words.most_common(12)))
```

```python

words_features = list(all_words.keys())[:1640]
print(words_features)
```

FEATURE EXTRACTION

```python
def find_features(text):
  words = word_tokenize(text)
  features = {}
  for word in words_features:
    features[word] = (word in words)
    
  return features

features = find_features(processed[0])
for key, value in features.items():
  if value == True:
    print(key,value)


message = list(zip(processed,y))
seed = 1
np.random.seed = seed
np.random.shuffle(message)
featureset = [(find_features(text),label) for (text,label) in message]

print(message)


```

SPLITING DATA INTO TRAIN AND TEST


```python
np.random.shuffle(message)



from sklearn import  model_selection
train,test = model_selection.train_test_split(featureset,test_size = 0.2,random_state = seed)


print('x_train; {}'.format(len(train)))
print('x_test; {}'.format(len(test)))
```
